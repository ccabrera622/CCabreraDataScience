```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv('company_sales_data.csv')
sales
```

Exercise 1: Read Total profit of all months and show it using a line plot

-Total profit data provided for each month. Generated line plot must include the following properties:

X label name = Month Number

Y label name = Total profit

```python
plt.plot(sales.month_number,sales.total_profit)

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Monthly Profit')
plt.ylim(0)
```

Exercise Question 2: Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

Line Style dotted and Line-color should be red

Show legend at the lower right location.

X label name = Month Number

Y label name = Sold units number

Add a circle marker.

Line marker color as read

Line width should be 3

```python
plt.plot(sales.month_number,sales.total_profit,':ro',label = 'profit', markersize=15, linewidth=3,
         markerfacecolor='white',
         markeredgecolor='red',
         markeredgewidth=2
)
plt.legend(loc = 'lower right')

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Monthly Profit')
plt.ylim(0)
```

Exercise Question 3: Read all product sales data and show it using a multiline plot
    
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
plt.plot(sales.month_number,sales.facecream, label = 'Face Cream')
plt.plot(sales.month_number,sales.facewash, label = 'Face Wash')
plt.plot(sales.month_number,sales.toothpaste, label = 'Toothpaste')
plt.plot(sales.month_number,sales.bathingsoap, label = 'Bathing Soap')
plt.plot(sales.month_number,sales.shampoo, label = 'Shampoo')
#plt.plot(sales.month_number,sales.moisturizer, label = 'Moisturizer')

plt.legend(loc = 'upper left', )

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Product Sales')
plt.ylim(0)
```

Exercise Question 4: Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should .

```python
plt.scatter(x = sales.month_number, y = sales.toothpaste, s = 50, c = 'green')

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Toothpaste Sold')
plt.ylim(0)

plt.xticks(range(13))

plt.grid(linestyle = '-')
```

Exercise Question 5: Read face cream and face wash product sales data and show it using the bar chart
    
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
br1 = np.arange(1,13) 
br2 = [x + 0.3 for x in br1]
plt.bar(br1,sales.facecream, label = 'Face Cream',width=0.3)
plt.bar(br2,sales.facewash, label = 'Face Wash',width=0.3)
plt.xticks(range(1,13))
plt.legend(loc = 'upper right')
```
